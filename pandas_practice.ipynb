```python
# --- PANDAS STRING METHODS CHEATSHEET ---

# 1. CASE TRANSFORMATION
df['col'].str.lower()      # Convert to lowercase
df['col'].str.upper()      # Convert to uppercase
df['col'].str.title()      # Convert to Title Case
df['col'].str.capitalize() # Capitalize first letter only

# 2. CLEANING & REPLACING
df['col'].str.strip()      # Remove leading and trailing whitespace
df['col'].str.lstrip()     # Remove whitespace on the left
df['col'].str.rstrip()     # Remove whitespace on the right
df['col'].str.replace('old', 'new') # Replace text (supports regex)

# 3. FILTERING & SEARCHING (Returns Boolean True/False)
df[df['col'].str.contains('keyword')]   # Check if string contains a substring
df[df['col'].str.startswith('prefix')]  # Check if string starts with prefix
df[df['col'].str.endswith('suffix')]    # Check if string ends with suffix
df[df['col'].str.match(r'^Regex')]      # Check if matches a regex pattern

# 4. SPLITTING & EXTRACTING
df['col'].str.split(' ')            # Split into a list of strings
df['col'].str.split(' ', expand=True) # Split into multiple NEW columns
df['col'].str.get(0)                # Get the element at index 0 (useful after split)
df['col'].str.extract(r'(\d+)')     # Extract first numeric match into a new column

# 5. INFORMATION & SLICING
df['col'].str.len()                 # Get length of each string
df['col'].str.slice(0, 5)           # Slice string (same as [0:5])
df['col'].str.count('a')            # Count occurrences of a character/pattern

# 6. PADDING & ENCODING
df['col'].str.zfill(5)              # Pad with zeros (e.g., '1' becomes '00001')
df['col'].str.pad(width=10, side='left', fillchar='-') # Custom padding
```

##### datetime operations 

```python
df['date_col'] = pd.to_datetime(df['col'])

# Assuming df['date_col'] is already converted via pd.to_datetime()

# 1. Extraction
df['year']  = df['date_col'].dt.year
df['month'] = df['date_col'].dt.month
df['day']   = df['date_col'].dt.day
df['hour']  = df['date_col'].dt.hour

# 2. Names and Labels
df['day_name']   = df['date_col'].dt.day_name()   # 'Monday', 'Tuesday'
df['month_name'] = df['date_col'].dt.month_name() # 'January'

# 3. Period/Frequency Logic
df['quarter']    = df['date_col'].dt.quarter      # 1, 2, 3, or 4
df['is_weekend'] = df['date_col'].dt.dayofweek > 4 # True for Sat/Sun
df['is_leap']    = df['date_col'].dt.is_leap_year

# 4. Formatting
df['formatted'] = df['date_col'].dt.strftime('%Y-%m') # Custom strings (e.g., '2023-12')
```

##### groupby

```python
# MATH - Basic Aggregation

df.groupby('dept')['salary'].sum()     # Total
df.groupby('dept')['salary'].mean()    # Average
df.groupby('dept')['salary'].median()  # Middle value
df.groupby('dept')['salary'].std()     # Standard Deviation
df.groupby('dept')['salary'].var()     # Variance

# COUNTS & EXTREMES
df.groupby('dept')['id'].count()       # Count non-null values
df.groupby('dept')['id'].nunique()     # Count unique values
df.groupby('dept')['salary'].min()     # Minimum
df.groupby('dept')['salary'].max()     # Maximum
df.groupby('dept')['salary'].first()   # First row in group
df.groupby('dept')['salary'].last()    # Last row in group

# Advanced Aggregation
# Multiple functions at once
df.groupby('dept')['salary'].agg(['mean', 'sum', 'count'])

# Named Aggregation (Best for clean columns)
df.groupby('dept').agg(
    avg_pay=('salary', 'mean'),
    max_exp=('experience', 'max'),
    staff_count=('id', 'count')
)

# Transformation & Filtration
# 1. TRANSFORMATION
# Returns a Series the SAME LENGTH as the original df.
# Useful for: "What is this person's salary vs the dept average?"
df['dept_avg'] = df.groupby('dept')['salary'].transform('mean')

# 2. FILTRATION
# Discards entire groups based on a condition.
# Keeps only depts with more than 10 employees:
df.groupby('dept').filter(lambda x: len(x) > 10)

# 3. APPLY
# The "Swiss Army Knife" - can return anything.
df.groupby('dept').apply(lambda x: x.nlargest(2, 'salary'))



# Rank employees by salary within their own department
df['dept_rank'] = df.groupby('department')['salary'].rank(method='dense', ascending=False)

```
